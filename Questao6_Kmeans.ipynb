#### Letra A: inicializando em $m_1 = (0,0,0)$, $m_2 = (1,1,1)$, $m_3 = (-1,0,2)$.

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X = np.array([
    [-7.82, -4.58, -3.97],
    [-6.68, 3.16, 2.71],
    [4.36, -2.19, 2.09],
    [6.72, 0.88, 2.80],
    [-8.64, 3.06, 3.50],
    [-6.87,0.57,-5.45 ],
    [4.47,-2.62,5.76],
    [6.73,-2.01,4.18],
    [-7.71,2.34,-6.33],
    [-6.91,-0.49,-5.68],
    [6.18,2.81,5.82],
    [6.72,-0.93,-4.04],
    [-6.25,-0.26,0.56],
    [-6.94,-1.22,1.13],
    [8.09,0.20,2.25],
    [6.81,0.17,-4.15],
    [-5.19,4.24,4.04],
    [-6.38,-1.74,1.43],
    [4.08,1.30,5.33],
    [6.27,0.93,-2.78],
])

begin = np.array([[0,0,0],[1,1,1],[-1,0,2]])

kmeans = KMeans(n_clusters=3, random_state=0, init=begin, max_iter=10000,tol=1e-5).fit(X)
kmeans.labels_

#kmeans.predict([[0, 0], [4, 4]])

c = kmeans.cluster_centers_ #each line is a cluster
print(c)
l = kmeans.labels_
print(l)


fig1 = plt.figure()
a1 = Axes3D(fig1)
a1.plot(X[l==0,0], X[l==0,1],  X[l==0,2] ,'o')
a1.plot(X[l==1,0], X[l==1,1],  X[l==1,2]  ,'o')
a1.plot(X[l==2,0], X[l==2,1],  X[l==2,2]  ,'o')
a1.plot(c[0, :], c[1, :], c[2, :] ,marker='P')

#a1.scatter()
a1.view_init(10,75)

plt.show()
```
